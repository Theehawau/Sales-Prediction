```python
# Get Data from DVC remote
import dvc.api
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import mlflow
import mlflow.sklearn
```

Function to get data from DVC remote:

```python
def data_from_DVC(path, repo, version):
    data_url = dvc.api.get_url(
        path=path,
        repo=repo,
        rev=version
    )
    return pd.read_csv(data_url, sep=',')
```

```python
store_db = data_from_DVC('data/store.csv','git@github.com:Theehawau/Sales-Prediction.git','v1')
store_db.head(2)
```

```python
# Read the train data from remote repo
train_db = data_from_DVC('data/train.csv','git@github.com:Theehawau/Sales-Prediction.git','v1')
train_db.head(2)
```

Filling null values for store dataset:

```python
store_db['Promo2SinceWeek'] = store_db['Promo2SinceWeek'].fillna(0)
store_db['Promo2SinceYear'] = store_db['Promo2SinceYear'].fillna(0)
store_db['PromoInterval'] = store_db['PromoInterval'].fillna(0)
store_db['CompetitionDistance'] =store_db['CompetitionDistance'].fillna(0)
```

Changing data types:
1. Converting Object date to datetime format
2. Converting float to int

```python
def to_datetime(df, col, format):
    df[col] = pd.to_datetime(df[col],format=format)
```

```python
to_datetime(test_db, 'Date', '%Y-%m-%d')
```

```python
to_datetime(train_db,'Date', '%Y-%m-%d')
```

```python
store_db = store_db.astype({'Promo2SinceYear':int})
```

```python
train_db = train_db.astype({'StateHoliday':str})
```

Mergin Store and Train dataset:

```python
db = pd.merge(left=train_db,right=store_db,on='Store',how='inner')
db.shape
```
